```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')

import pandas as pd
data = pd.read_csv("data\GlobalTemperatures.csv", header=0, index_col=0, parse_dates=True, date_format="%Y/%d/%m")


series = pd.DataFrame()
series['date'] = data.index
series['year'] = pd.DatetimeIndex(data.index).year
series['temp'] = data["LandAverageTemperature"].values

series = series[series['year'] >= 1800]

fig, ax = plt.subplots()
fig.set_figwidth(20)
ax.grid(True)
ax.grid(which='major', color='#DDDDDD', linewidth=0.5, alpha=0.2)
ax.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5,alpha=0.2)
ax.minorticks_on()
ax.axhline(y=0.0, color='w', linestyle='-', alpha=0.7)
ax.set_ylim([-5,17])


for i in range(0,12,1):
    offset = i
    period = 12
    X = series['date'].index[offset::period]
    Y = series['temp'].values[offset::period]
    ax.plot(X, Y, color=[np.sin(offset/10)**2, 0,np.cos(offset/10)**2], alpha=0.6)

offset = 2
period = 12
X = series['date'].index[offset::period]
Y = series['temp'].values[offset::period]
ax.plot(X, Y, color=[0.1, 0.9, 0.2])

ax.set_xticks(series['year'].index[::int(12*10)])
ax.set_xticklabels(list(map(str,series['year'].values[::int(12*10)])))

Y = Y.mean()
ax.axhline(y=Y, color=[0.9, 0.1, 0.2], linestyle=':')




ax.axhline(y=0, color=[0.0, 0.0, 0.0], linestyle='-', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')

import pandas as pd
data = pd.read_csv("data\GlobalTemperatures.csv", header=0, index_col=0, parse_dates=True, date_format="%Y/%d/%m")


series = pd.DataFrame()
series['date'] = data.index
series['year'] = pd.DatetimeIndex(data.index).year
series['temp'] = data["LandAverageTemperature"].values

series = series[series['year'] >= 1800]

fig, ax = plt.subplots()
fig.set_figwidth(20)
ax.grid(True)
ax.grid(which='major', color='#DDDDDD', linewidth=0.5, alpha=0.2)
ax.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5,alpha=0.2)
ax.minorticks_on()
ax.axhline(y=0.0, color='w', linestyle='-', alpha=0.7)
ax.set_ylim([-5,17])


temps = pd.DataFrame(series['temp'].values)
window = temps.rolling(window=48)

mean_temps = window.mean()
min_temps = window.min()
max_temps = window.max()

offset = 0
period = 1
X = series['year'].index
Y = mean_temps.values[offset::period]
ax.plot(X, Y, color=[0.9, 0.1, 0.9])

X = series['year'].index
Y = min_temps.values[offset::period]
ax.plot(X, Y, color=[0.8, 0.3, 0.2])

X = series['year'].index
Y = max_temps.values[offset::period]
ax.plot(X, Y, color=[0.3, 0.2, 0.8])



offset = 0
period = 1
X = series['date'].index[offset::period]
Y = series['temp'].values[offset::period]
ax.plot(X, Y, color=[0.1, 0.1, 0.1], alpha=0.3)



Y = Y.mean()
ax.axhline(y=Y, color=[0.9, 0.1, 0.2], linestyle=':')




ax.axhline(y=0, color=[0.0, 0.0, 0.0], linestyle='-', alpha=0.5)

ax.set_xticks(series['year'].index[::int(12*10)])
ax.set_xticklabels(list(map(str,series['year'].values[::int(12*10)])))

plt.show()
```
