```python
#Gloabal_Temperature_Dense

import pandas as pd

temeperature_anomalies_series = pd.read_csv("data\Gloabal_Temperature_Dense.csv", header=0, index_col=0, parse_dates=True)
#data = data.loc[(data.index > '1920')]
#data = data.iloc[::12]

temeperature_anomalies_series.tail(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()

fig.set_figwidth(15)

ax.grid(True)
# Show the major grid and style it slightly.
ax.grid(which='major', color='#DDDDDD', linewidth=0.5, alpha=0.2)
# Show the minor grid as well. Style it in very light gray as a thin,
# dotted line.
ax.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5,alpha=0.2)
# Make the minor ticks and gridlines show.
ax.minorticks_on()
X = temeperature_anomalies_series.index.values
Y = temeperature_anomalies_series["Anomaly"].values
#plt.xticks(X)


ax.axhline(y=0.0, color='w', linestyle='-', alpha=0.7)
ax.plot(X, Y, "-", color=[0.3,0.6,0.9])
ax.plot(X, Y, "+", color="w")
```
