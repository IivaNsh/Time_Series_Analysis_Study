```python
import pandas as pd
```

```python
data = pd.read_csv("data\GlobalTemperatures.csv", header=0, index_col=0, parse_dates=True)
```

```python
data = data.loc[(data.index > '1920')]
data = data.iloc[::12]
```

```python
data.tail(10)
```

```python
average_temperature_values_global = data["LandAverageTemperature"].values
average_temperature_values_global.size
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.grid(True)
# Show the major grid and style it slightly.
ax.grid(which='major', color='#DDDDDD', linewidth=0.5)
# Show the minor grid as well. Style it in very light gray as a thin,
# dotted line.
ax.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
# Make the minor ticks and gridlines show.
ax.minorticks_on()
ax.plot(data.index.values, average_temperature_values_global, "-")
```
